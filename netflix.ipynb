```python
df = df.drop(1704)
```

```python
df.shape
```

```python
# Remove unnecessary characters
```

```python
df['genres'] = df['genres'].str.replace('\'', '', regex=True)
```

```python
df['genres'] = df['genres'].str.replace('\[', '', regex=True)
```

```python
df['genres'] = df['genres'].str.replace('\]', '', regex=True)
```

```python
df['genres']
```

```python
df['production_countries'] = df['production_countries'].str.replace('\'', '', regex=True)
```

```python
df['production_countries'] = df['production_countries'].str.replace('\[', '', regex=True)
```

```python
df['production_countries'] = df['production_countries'].str.replace('\]', '', regex=True)
```

```python
df['production_countries']
```

```python
df['production_countries']
```

```python
# Convert float seasons to int
```

```python
df['seasons'] = df['seasons'].astype(pd.Int32Dtype())
```

```python
df['imdb_votes'] = df['imdb_votes'].astype(pd.Int32Dtype())
```

```python
df.head()
```

```python
# Round out tmdb score to provide equal comparison 
```

```python
df['tmdb_popularity'] = df['tmdb_popularity'].round(2) 
```

```python
df['tmdb_score'] = df['tmdb_score'].round(1) 
```

```python
df.to_csv('titles_final.csv', index=False)
```
